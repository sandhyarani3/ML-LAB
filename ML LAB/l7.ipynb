```python
import pandas as pd
df_train=pd.read_csv("Week7_train.csv")
df_test=pd.read_csv("Week7_test.csv")
```

```python
df_train=df_train.sample(frac=0.1,random_state=42)
```

```python
X_train=df_train.drop(columns=['label'])
y_train=df_train['label']
```

```python
X_train
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
for i in range(5):
    plt.imshow(X_train.iloc[i].values.reshape(28,28),cmap="gray")
    plt.title(y_train.iloc[i])
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X_train,y_train,test_size=0.25,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=5,metric="euclidean")
KNN.fit(X_train,y_train)
```

```python
df_test
```

```python
y_pred=KNN.predict(df_test)
```

```python
y_pred
```

```python
KNN=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
KNN.fit(X_train,y_train)
```

```python
y_pred=KNN.predict(X_valid)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_valid,y_pred)
```

```python
accuracy_score(y_valid,y_pred)
```

```python
classification_report(y_valid,y_pred)
```

```python
for i in range(5):
    plt.imshow(X_valid.iloc[i].values.reshape(28,28),cmap="gray")
    plt.title(y_valid.iloc[i])
    plt.show()
```

```python
y_pred=KNN.predict(X_valid)
y_pred
```

```python
cm=confusion_matrix(y_valid,y_pred)
import seaborn as sns
sns.heatmap(cm,fmt='d',annot=True)
plt.show()
```
