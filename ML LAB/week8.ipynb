```python
X=df.drop(columns=["Species"])
y=df["Species"]
```

```python
df.select_dtypes(include=['number'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Species"]=le.fit_transform(df["Species"])
```

```python
y=df["Species"]
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel="linear",C=1)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.svm import SVC
model=SVC(kernel="linear",C=0.1)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.model_selection import KFold,cross_val_score
model=SVC(kernel="linear",C=1)
kf=KFold(n_splits=5,shuffle=True,random_state=42)
scores=cross_val_score(model,X,y,cv=kf)
scores
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
SVC=SVC(kernel="linear")
param_grid={'C':[0.1,1,10,100,1000]}
gridsearch=GridSearchCV(SVC,param_grid,cv=kf,scoring="accuracy")
gridsearch.fit(X_train,y_train)
```

```python
gridsearch.best_params_['C']
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
```

```python
y_pred=tree.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
res_df=pd.DataFrame(gridsearch.cv_results_)
```

```python
res_df
```

```python
C_values=[0.1,1,10,100,1000]
train_scores=[]
test_scores=[]
from sklearn.svm import SVC
for c in C_values:
    model=SVC(kernel="linear",C=c)
    model.fit(X_train,y_train)
    train_scores.append(model.score(X_train,y_train))
    test_scores.append(model.score(X_test,y_test))
plt.plot(C_values,train_scores,marker='o',label="Train scores")
plt.plot(C_values,test_scores,marker='s',label="Test scores")
plt.xscale('log')
plt.show()
```
